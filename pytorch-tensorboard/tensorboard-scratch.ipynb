```python
import torch
import torchvision
from torch.utils.tensorboard import SummaryWriter
from torchvision import datasets, transforms

import numpy as np
```

```python
writer = SummaryWriter()

'''
MNIST PIL 데이터를 pytorch Tensor로 변환합니다.
* transfroms.Compose(): transform 파라미터를 행렬 형태로 추가합니다.
* transforms.ToTensor(): pytorch Tensor로 변환합니다.
* transfroms.Normalize(): Tensor 변환한 이미지를 정규화 합니다. 
'''
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
# MNIST 데이터 객체를 선언합니다.
trainset = datasets.MNIST('mnist_train', train=True, download=True, transform=transform)
# 데이터를 불러옵니다.
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True, num_workers=2)

model = torchvision.models.resnet50(False)
model.conv1 = torch.nn.Conv2d(1, 64, kernel_size=3, stride=2, padding=0, bias=False)
images, labels = next(iter(trainloader))

grid = torchvision.utils.make_grid(images)
writer.add_image('images', grid, 0)
# writer.add_graph(model, images)
writer.close()
```

```python
for n_iter in range(100):
    writer.add_scalar('Loss/train', np.random.random(), n_iter)
    writer.add_scalar('Loss/test', np.random.random(), n_iter)
    writer.add_scalar('Accuracy/train', np.random.random(), n_iter)
    writer.add_scalar('Accuracy/test', np.random.random(), n_iter)
```
